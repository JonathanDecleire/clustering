```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
from sklearn.cluster import AgglomerativeClustering
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.metrics import adjusted_rand_score
```

```python
df = pd.read_csv('chipotle_stores.csv')
```

```python
import scipy.cluster.hierarchy as sch
X = df.loc[:,['longitude','latitude']]

plt.figure(figsize=(18,10))
dendrogram = sch.dendrogram(sch.linkage(X, method  = "ward"), p=30, truncate_mode='lastp')

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
kmeans = KMeans(init="random", n_clusters=23, n_init=10, max_iter=300, random_state=42)
```

```python
coor = df[['latitude', 'longitude']]
```

```python
kmeans.fit(coor)
```

```python
# The lowest SSE value
kmeans.inertia_
```

```python
# Final locations of the centroid
kmeans.cluster_centers_
```

```python
# The number of iterations required to converge
kmeans.n_iter_
```

```python
kmeans_kwargs = {"init": "random",
                    "n_init": 10,
                     "max_iter": 300,
                      "random_state": 42,}

# A list holds the SSE values for each k
sse = []
for k in range(1, 30):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(coor)
    sse.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,6))
plt.style.use("fivethirtyeight")
plt.plot(range(1, 30), sse)
plt.xticks(range(1, 30))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(range(1, 30), sse, curve="convex", direction="decreasing")
kl.elbow
```

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 30):
     kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
     kmeans.fit(coor)
     score = silhouette_score(coor, kmeans.labels_)
     silhouette_coefficients.append(score)
```

```python
len(silhouette_coefficients)
```

```python
plt.figure(figsize=(8,6))
plt.style.use("fivethirtyeight")
plt.plot(range(2, 30), silhouette_coefficients)
plt.xticks(range(2, 30))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
# Instantiate k-means and dbscan algorithms
kmeans = KMeans(n_clusters=19)
dbscan = DBSCAN(eps=0.3)

# Fit the algorithms to the features
kmeans.fit(coor)
dbscan.fit(coor)

# Compute the silhouette scores for each algorithm
kmeans_silhouette = silhouette_score(coor, kmeans.labels_).round(2)
dbscan_silhouette = silhouette_score(coor, dbscan.labels_).round (2)
```

```python
kmeans_silhouette
```

```python
dbscan_silhouette
```

```python
kmeans.labels_
```

```python
# Plot the data and cluster silhouette comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 6), sharex=True, sharey=True)
fig.suptitle(f"Clustering Algorithm Comparison: Crescents", fontsize=16)
fte_colors = {0: "#008fd5",
                1: "#fc4f30"}
# The k-means plot 'latitude', 'longitude'
km_colors = [coor[label] for label in kmeans.labels_]
ax1.scatter(coor[:, 0], coor[:, 1], c=km_colors)
ax1.set_title(f"k-means\nSilhouette: {kmeans_silhouette}", fontdict={"fontsize": 12})

# The dbscan plot
db_colors = [coor[label] for label in dbscan.labels_]
ax2.scatter(coor[:, 0], coor[:, 1], c=db_colors)
ax2.set_title(f"DBSCAN\nSilhouette: {dbscan_silhouette}", fontdict={"fontsize": 12})
plt.show()
```
